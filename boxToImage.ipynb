# Function Returning cropped image to use further

```python
def crop(img,dict):
    (im_width, im_height) = image.size
    score = max(dict["detection_scores"])
    (ymin,xmin,ymax,xmax) = dict["detection_boxes"][0]
    (left, right, top, bottom) = (xmin * im_width, xmax * im_width,
                                  ymin * im_height, ymax * im_height)
    fig , ax = plt.subplots()
    ax.imshow(image)
    ax.set_ylim((bottom,top))
    ax.set_xlim((left,right))
    
    width = right - left
    height = bottom - top
    
    (x,y) = (left,top)
    
    area = (x,y,width,height)
    
    image_2 = image.crop(area)
    
    plt.imshow(image_2)
    
    return image_2
```
